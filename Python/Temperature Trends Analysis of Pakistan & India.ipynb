## Libraries used & Data Slicing and Segregation

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset1=pd.read_csv("Tempreture_1901_2016_Pakistan.csv")
array1=dataset1.values
paktemperature=array1[:,0]
pakyear=array1[:,1]
pakmonth=array1[:,2]

pakyear=list(dict.fromkeys(pakyear))
pakmonth=list(dict.fromkeys(pakmonth))

dataset2=pd.read_csv("Temperature 1901-2017 India.csv")
array2=dataset2.values
indtemperature=array2[:,0]
indyear=array2[:,1]
indmonth=array2[:,2]

indyear=list(dict.fromkeys(indyear))
indmonth=list(dict.fromkeys(indmonth))
```

## Calculating Average Temperature For Each Year Pakistan & India

```python
start=0
end=12
avg=0
pakaveragetemp=[] 
for x in range(0,int(len(paktemperature)/12)):
    for y in range (start, end):
        avg=avg+paktemperature[y]

    
    pakaveragetemp.append(avg/12)
    avg=0
    start=start+12
    end=end+12   
```

```python
start=0
end=12
avg=0
indaveragetemp=[] 
for x in range(0,int(len(indtemperature)/12)):
    for y in range (start, end):
        avg=avg+indtemperature[y]

    
    indaveragetemp.append(avg/12)
    avg=0
    start=start+12
    end=end+12 
```

## Visualizing Average Temperature Each Year of Pakistan & India

```python
plt.figure(figsize=(15,10))
plt.plot(pakyear,pakaveragetemp, color="green", marker="o")
plt.title('Average Temperature of Pakistan Over Each Year')
plt.xlabel('Years')
plt.ylabel('Average Temperature')
plt.grid(True)
plt.show()

plt.figure(figsize=(15,10))
plt.plot(indyear,indaveragetemp, color="blue", marker="o")
plt.title('Average Temperature of India Over Each Year')
plt.xlabel('Years')
plt.ylabel('Average Temperature')
plt.grid(True)
plt.show()


plt.figure(figsize=(15,10))
plt.plot(pakyear,pakaveragetemp, color="green", label='Pakistan' , marker="o")
plt.plot(indyear,indaveragetemp, color="blue", label='India' , marker="o")
plt.title('Comparison of Average Temperature of Pakistan vs India')
plt.xlabel('Years')
plt.ylabel('Average Temperature')
plt.grid(True)
plt.legend()
plt.show()

```

## Calculating Temperatures For Summer (May & June) Pakistan & India

```python
pakmay=[]
pakjun=[]
req1=4
req2=5
for x in range(int(len(paktemperature)/12)):
    pakmay.append(paktemperature[req1])
    pakjun.append(paktemperature[req2])
    
    req1=req1+12
    req2=req2+12
```

```python
indmay=[]
indjun=[]
req1=4
req2=5
for x in range(int(len(indtemperature)/12)):
    indmay.append(indtemperature[req1])
    indjun.append(indtemperature[req2])
    
    req1=req1+12
    req2=req2+12
```

```python
paksumavg=[]
indsumavg=[]

for x in range(len(pakmay)):
    paksumavg.append((pakmay[x] + pakjun[x])/2)
    
for y in range(len(indmay)):
    indsumavg.append((indmay[y] + indjun[y])/2)
```

## Visualizing Temperature Trends For Summer (May-June) Pakistan & India

```python
plt.figure(figsize=(15,10))
plt.plot(pakyear, pakmay, color='Green', label='Pakistan', marker='o')
plt.plot(indyear, indmay, color='Blue', label='India', marker='o')
plt.title('Temperature Trends For May Pakistan vs India')
plt.xlabel('Years')
plt.ylabel('Temperature In Celsius')
plt.grid(True)
plt.legend()
plt.show

plt.figure(figsize=(15,10))
plt.plot(pakyear, pakjun, color='Green', label='Pakistan', marker='o')
plt.plot(indyear, indjun, color='Blue', label='India', marker='o')
plt.title('Temperature Trends For June Pakistan vs India')
plt.xlabel('Years')
plt.ylabel('Temperature In Celsius')
plt.grid(True)
plt.legend()
plt.show

plt.figure(figsize=(15,10))
plt.plot(pakyear, paksumavg, color='Green', label='Pakistan', marker='o')
plt.plot(indyear, indsumavg, color='Blue', label='India', marker='o')
plt.title('Average Temperature Trends For Summer May-June Pakistan vs India')
plt.xlabel('Years')
plt.ylabel('Temperature In Celsius')
plt.grid(True)
plt.legend()
plt.show
```
